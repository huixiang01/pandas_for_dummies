```python
# This notebook teaches you to use merge, sort, reset_index and fillna.
# First, import pandas

import pandas as pd
```

```python
# Merge
# let's do this

zoo_eats = pd.DataFrame([['elephant','vegetables'], ['tiger','meat'], ['kangaroo','vegetables'], ['zebra','vegetables'], ['giraffe','vegetables']], columns=['animal', 'food'])
zoo_eats
```

```python
# and this

zoo = pd.read_csv("zoo.csv")
zoo
```

```python
# If we merge them together, it becomes this:

zoo.merge(zoo_eats)

# The corresponding food would be sorted into the coresponding animal
```

```python
# If you do it another way around, the only difference would be the order 

zoo_eats.merge(zoo)
```

```python
# "Wait a minute, where're lion and giraffe in the picture?"
# Okay then, how about putting "NaN" value on lion and giraffe instead.

zoo.merge(zoo_eats, how = 'outer')
```

```python
# "I just want specific "NaN" to be shown(e.g. I don't want giraffe but I want lion to appear.)"
# Sure.

zoo.merge(zoo_eats, how = 'left')
```

```python
# By using different how parameter, you choose the things you want!

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image = mpimg.imread("./Images/join-or-merge-in-python-pandas-1.png")
                    
plt.imshow(image)
plt.show()
```

```python
# "Uhhh... NaN so ugly."
# You can fill it with something else... like "unknown"

zoo.merge(zoo_eats, how = 'left').fillna('unknown')
```

```python
# In some cases, pandas needs the common variable to base on: 
# In this case, both are based on "animals"

zoo.merge(zoo_eats, how = 'left', left_on = 'animal', right_on = 'animal')
```

```python
# Sorting...
# To sort values:
zoo.sort_values('water_need')
```

```python
# And to prioritise what comes first, you can switch around/add/substract values.
zoo.sort_values(by = ['animal', 'water_need'])
```

```python
#If you want all to increase/decrease:
zoo.sort_values(by = ['water_need'], ascending = False)
zoo.sort_values(by = ['water_need'], ascending = True)
```

```python
# "Okay, sorted already. And I wanna make it neat."

# Can just reset_index()
zoo.sort_values(by = ['water_need'], ascending = False).reset_index()
```

```python
# You can also drop the initial index
zoo.sort_values(by = ['water_need'], ascending = False).reset_index(drop = True)
```
